```python
import numpy as np
import pandas as pd
import folium


m = folium.Map(
    location = [36.4989, 127.9606],
    zoom_start = 8
)

o = pd.read_csv('rest-area3.csv', encoding='CP949')

쉼터유 = folium.FeatureGroup(name='쉼터유').add_to(m)
쉼터무 = folium.FeatureGroup(name='쉼터무').add_to(m)

for i in range(o.shape[0]):
    if o.iloc[i]["쉼터유무"] == 'Y':
        folium.Marker(
            [o.iloc[i]['위도'], o.iloc[i]['경도']],
            popup = f'<div style="width: 120px;"><strong>{o.iloc[i]["휴게소명"]}</strong><br>\
                        전기차충전소유무 : <strong>{o.iloc[i]["전기차충전소유무"]}</strong><br>\
                        세차장 : <strong>{o.iloc[i]["세차장"]}</strong><br>\
                        휴게소대표음식명 : <strong>{o.iloc[i]["휴게소대표음식명"]}</strong><br>\
                        쉼터유무 : <strong>{o.iloc[i]["쉼터유무"]}</strong></div>',
            tooltip = o.iloc[i]["휴게소명"]
        ).add_to(쉼터유)
    else:
        folium.Marker(
            [o.iloc[i]['위도'], o.iloc[i]['경도']],
            popup = f'<div style="width: 120px;"><strong>{o.iloc[i]["휴게소명"]}</strong><br>\
                        전기차충전소유무 : <strong>{o.iloc[i]["전기차충전소유무"]}</strong><br>\
                        세차장 : <strong>{o.iloc[i]["세차장"]}</strong><br>\
                        휴게소대표음식명 : <strong>{o.iloc[i]["휴게소대표음식명"]}</strong><br>\
                        쉼터유무 : <strong>{o.iloc[i]["쉼터유무"]}</strong></div>',
            tooltip = o.iloc[i]["휴게소명"]
        ).add_to(쉼터무)

folium.LayerControl(collapsed=False).add_to(m)

m
```

```python
<py-config>
    packages = [
        "folium",
        "pandas"
    ]
    plugins = [
        "https://pyscript.net/latest/plugins/python/py_tutor.py"
    ]
</py-config>

<py-script>
    import folium
    import json
    import pandas as pd

    from pyodide.http import open_url

    url = "https://raw.githubusercontent.com/python-visualization/folium/master/examples/data"
    country_geo = f"{url}/world-countries.json"
    country_population = f"{url}/world_population.csv"
    country_data = pd.read_csv(open_url(country_population))
    geo_json = json.loads(open_url(country_geo).read())

    m = folium.Map(location=[35.9078, 127.7669], zoom_start=7)

    folium.Choropleth(
        geo_data=geo_json,
        name="choropleth",
        data=country_data,
        columns=["Country", "Population"],
        key_on="feature.properties.name",
        fill_color="YlOrRd",
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name="Population",
    ).add_to(m)

    folium.LayerControl().add_to(m)

    display(m, target="folium")
</py-script>
```
